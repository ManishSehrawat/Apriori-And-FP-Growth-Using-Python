# Turning the dataset into 0 or 1 type----(if the item was bought in a transaction it will be represented by 1 else 0)

```python
encoded_vals = []
for index, row in filt.iterrows():
    labels = {}
    uncommons = list(set(unique_items) - set(row))
    commons = list(set(unique_items).intersection(row))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
encoded_vals[0]
ohe_df = pd.DataFrame(encoded_vals)
ohe_df
```

# Applying Apriori Algorithm
## Setting Minimum Support Count = 0.04% and Confidence = 70%

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_items = apriori(ohe_df, min_support=0.004, use_colnames=True, verbose=1)
freq_items.head(10)
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.70)
rules.sort_values('confidence', axis=0, ascending=False, inplace=True)
```

# Top 10 Associations:

```python
#Sort the rules 
rules_fp = rules.sort_values(['confidence'], ascending =[False])
#From Frozenset to string
rules_fp["antecedents"] = rules_fp["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
rules_fp["consequents"] = rules_fp["consequents"].apply(lambda x: list(x)[0]).astype("unicode")
rules_fp['support'] = rules_fp['support'].apply(lambda value: round(value, 2))
rules_fp['confidence'] = rules_fp['confidence'].apply(lambda value: round(value, 2))
rules_fp['lift'] = rules_fp['lift'].apply(lambda value: round(value, 2))
rules_fp['leverage'] = rules_fp['leverage'].apply(lambda value: round(value, 2))
rules_fp['conviction'] = rules_fp['conviction'].apply(lambda value: round(value, 2))
pd.set_option('display.max_columns', 10)
rules_fp.head(10)
```

```python
rules_top_fp = rules_fp.iloc[0:4, :]
labels = ["=>".join(rule) for rule in rules_top_fp.iloc[:, 0:2].values]  # format {set1} => {set2} for labels
x = np.arange(len(labels))  # the label locations
width = 0.3
fig, ax = plt.subplots(figsize=(15,6))
rects1 = ax.bar(x - width/2, rules_top_fp['conviction'].values, width, label='conviction')
rects2 = ax.bar(x + width/2, rules_top_fp['confidence'].values, width, label='confidence')
rects3 = ax.bar(x + width*1.5, rules_top_fp['lift'].values, width, label='lift')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Metrics')
ax.set_title('Metrics for top ' + str(3) + ' rules evaluated with ' + "lift")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

mng = plt.get_current_fig_manager()
# mng.window.state('zoomed')  # maximize the screen in windows
plt.show()
```

# Applying FP - Growth Algorithm

```python
# Dropping Item(s) column
trasn = filt.drop('Item(s)',axis = 1)
```

### Making a list of all transactions

```python
transactions = []
for i in range(0, len(trasn)):
    transactions.append([str(trasn.values[i,j]) for j in range(0, 32)])
transactions
```

### Code to remove null values inside lists of transaction

```python
final = []
```

```python
for i in range(0,len(transactions)):
    no_nan = []
    for j in range(0,len(transactions[i])):
        if transactions[i][j]!='nan':
            no_nan.append(transactions[i][j])
    final.append(no_nan)
```

```python
# Final transaction list with no null values
final
```

## Setting Minimum Support count = 15 and Confidence = 80%

```python
import pyfpgrowth
```

```python
patterns = pyfpgrowth.find_frequent_patterns(final,15)
```

```python
patterns
```

```python
rules1 = pyfpgrowth.generate_association_rules(patterns, 0.80)
rules1
```

## Arranging the output into interpretable format

```python
antecedents = []
for k in rules1.keys():
    antecedents.append(k)
```

```python
consequents = []
for k, v in rules1.items():
    consequents.append(v[0])
```

```python
confidence = []
for k, v in rules1.items():
    confidence.append(v[1])
```

```python
data = {}
data = {'Antecedents':antecedents,'Consequents':consequents,'Confidence':confidence}
```

```python
rules_df = pd.DataFrame(data)
rules_df
```

## Function to calculate Support Count of consequent items for calculating the Lift

```python
def support_count(rhs):
    count = 0
    rhs = set(rhs)
    for j in final:
        j = set(j)
        if(rhs.issubset(j)):
            count = count + 1
    return count
```

```python
rhs_support = []
for i in rules_df["Consequents"]:
    a = support_count(i)
    rhs_support.append(a/len(final))
```

```python
rules_df['RHS_support'] = rhs_support
```

```python
rules_df['lift'] = rules_df['Confidence']/rules_df['RHS_support']
```

## Calculating The Conviction

```python
rules_df['Conviction'] = (1-rules_df['RHS_support'])/(1-rules_df['Confidence'])
```

```python
rules_df.sort_values('Confidence', axis=0, ascending=False, inplace=True)
rules_df
```
